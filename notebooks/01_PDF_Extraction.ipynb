## Build PDF Ingestion Pipeline that automates files through only Folder_path and then save them in a Dictionary:

```python
import os 
import pdfplumber
folder_path = r"C:\Users\riaze\Desktop\TechCorp-Financial-Enterprise-RAG\data\pdfs\Apple"
company_data = {}
for file in os.listdir(folder_path):
    full_path = os.path.join(folder_path, file)
    full_text = ""
    with pdfplumber.open(full_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            if text:
                full_text += text + "\n"
    company_data[file] = full_text
    print(file)
    print(len(full_text))
```

## What I Learned

- Read a single PDF using pdfplumber.
- Extract text page by page.
- Automate extraction for multiple PDFs using os.listdir().
- Create full file paths using os.path.join().
- Count extracted characters to verify the successful extraction.
- Save All Extractd Text with their File names in a Dictionary

```python
import os
import pdfplumber
import gc

# PDF Folders
main_folder = r"C:\Users\riaze\Desktop\TechCorp-Financial-Enterprise-RAG\data\pdfs"

# Output Folder
output_folder = r"C:\Users\riaze\Desktop\TechCorp-Financial-Enterprise-RAG\data\extracted_text"
os.makedirs(output_folder , exist_ok=True)
total_companies = 0
total_pdfs = 0
 # Loop through all company folders
for company in os.listdir(main_folder):
    company_path = os.path.join(main_folder, company)
    if not os.path.isdir(company_path):
        continue
    print(f"\n  Company: {company}")

    # Create company folder inside extracted_text
    company_output = os.path.join(output_folder, company)
    os.makedirs(company_output, exist_ok=True)
    company_pdf_count = 0 

    # Loop through Pdfs
    for file in os.listdir(company_path):
        if file.endswith(" .pdf"):
            pdf_path = os.path.join(company_path, file)
            full_text = ""
            try:
                with pdfplumber.open(pdf_path) as pdf:
                    for page in pdf.pages:
                        text = page.extract_text()
                        if text:
                            full_text += text + "\n"
                            # Create txt filename
                            txt_file = file.replace(".pdfs", ".txt")
                            save_path = os.path.join(company_output, txt_file)
                            with open(save_path, "w", encoding = "utf-8") as f:
                                f.write(full_text)
                                # Free RAM
                                del full_text
                                gc.collect()
                                company_pdf_count += 1
                                total_pdfs += 1
                                print(f" Done Saved: {txt_file}")
                        except Exception as e:
                                print(f" Wrong Error: {file}")
                                print(e)
                            total_companies += 1
                            print(f" Pdfs Processed: {company_pdf_count}")
                            print("\n=====================================")
                            print("Enterprise Data Ingestion Complete")
                            print(f" Total Companies : {total_companies}")
                            print(f" Total Pdfs   : {total_pdfs}")
```
