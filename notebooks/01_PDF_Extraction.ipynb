## Exraction of Full Text , Basic Cleaning and Debugging of Google_2021 Financial Report

```python
import pdfplumber
pdf_path = r"C:\Users\riaze\Desktop\TechCorp-Financial-Enterprise-RAG\data\pdfs\Google\Google_2021_10K.pdf"
full_text = ""
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            full_text += text + "\n"
print(len(pdf.pages))
print(len(full_text))
print(full_text[:1000])
## Basic Cleaning of PDF:
clean_text = " ".join(full_text.split())
print(clean_text[:1000])
print(len(clean_text))
```

## Build PDF Ingestion Pipeline that automates files through only Folder_path:

```python
import os 
import pdfplumber
folder_path = r"C:\Users\riaze\Desktop\TechCorp-Financial-Enterprise-RAG\data\pdfs\Apple"
for file in os.listdir(folder_path):
    full_path = os.path.join(folder_path, file)
    full_text = ""
    with pdfplumber.open(full_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            if text:
                full_text += text + "\n"
    print(file)
    print(len(full_text))
```
