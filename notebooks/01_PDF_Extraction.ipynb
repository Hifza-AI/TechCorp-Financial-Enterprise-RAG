## Exraction of Full Text and Basic Cleaning

```python
import pdfplumber
pdf_path = r"C:\Users\riaze\Desktop\TechCorp-Financial-Enterprise-RAG\data\pdfs\Paypal\PayPal _2018_10K.pdf"
full_text = ""
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            full_text += text + "\n"
print(len(pdf.pages))
print(len(full_text))
print(full_text[:1000])
## Basic Cleaning of PDF:
clean_text = " ".join(full_text.split())
print(clean_text[:1000])
print(len(clean_text))
```
