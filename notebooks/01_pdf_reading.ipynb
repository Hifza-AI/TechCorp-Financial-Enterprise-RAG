## Exraction of Full Text 

```python
import pdfplumber
pdf_path = r"C:\Users\riaze\Desktop\TechCorp-Financial-Enterprise-RAG\data\pdfs\Apple\Apple_2024_10K.pdf"
full_text = ""
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        if text:
            full_text += text + "\n"
print(full_text[:10000])
print(len(pdf.pages))
print(len(full_text))
```
